```python
import numpy as np
import pandas as pd
import seaborn as sns
from matplotlib import pyplot as plt
import csv
```

Uploading The Dataset

```python
LaborForce='https://raw.githubusercontent.com/JanBii/ATT-FINAL_PROJECT_ALDAY-MAGNAYE-MENDOZA/main/EAP_2WAP_SEX_AGE_RT_A-filtered-2023-05-25%20(2).csv'
```

Reading and Printing the Dataset

```python
Labor_Force = pd.read_csv('https://raw.githubusercontent.com/JanBii/ATT-FINAL_PROJECT_ALDAY-MAGNAYE-MENDOZA/main/EAP_2WAP_SEX_AGE_RT_A-filtered-2023-05-25%20(2).csv')
Labor_Force
```

Renaming Labor_Force header

```python
Labor_Force.rename(columns={'ref_area.label' : 'Country', 'indicator.label' : 'indicator.label' , 'source.label' : 'Source', 'sex.label' : 'Sex', 'classif1.label' : 'Age', 'time' : 'Year','obs_value' : 'Percentage', 'obs_status.label' : 'Status'}, inplace=True)
Labor_Force
```

Removes the rows that contains NULL values in Labor_Force

```python
Labor_Force.fillna(0 , inplace=True)

Labor_Force
```

```python
sns.set(style="ticks")
plt.figure(figsize=(20, 100))  # Increase the width of the graph

sns.catplot(x="Year", y="Age", hue="Sex", kind="box", data=Labor_Force)
plt.title('Age of Employee per Year')
plt.xlabel('Year')
plt.ylabel('Age')
plt.show()
```

Year of Employment per Country

```python
sns.countplot(y='Country',hue='Year',data=Labor_Force)
sns.set(rc={'figure.figsize':(10,10)})
plt.xticks(rotation=90)
plt.title('Year of Employment per Country')
plt.legend(loc=1)  # 1 is code for 'upper right'
```

Labor Force per Country

```python
country_percentage = (Labor_Force['Country'].value_counts() / Labor_Force['Country'].value_counts().sum()) * 100

# Create a bar plot
plt.figure(figsize=(12, 8))
country_percentage.plot(kind='bar')
plt.title('Percentage of Labor Force per Country')
plt.xlabel('Country')
plt.ylabel('Percentage')
plt.xticks(rotation=45)
plt.show()
```

Sex Employed per Country

```python
plt.figure(figsize=(12, 8))
sns.countplot(x='Country', hue='Sex', data=Labor_Force)
plt.title('Sex per Country')
plt.xlabel('Country')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.legend(title='Sex')
plt.show()
```

Distribution of Age with each Country

```python
country_age_percentage = Labor_Force.groupby(['Country', 'Age']).size() / len(Labor_Force) * 100

# Reset the index to have Country, Age, and Percentage as columns
country_age_percentage = country_age_percentage.reset_index(name='Percentage')

# Create a scatter plot
plt.figure(figsize=(12, 8))
sns.scatterplot(data=country_age_percentage, x='Country', y='Age', size='Percentage', hue='Percentage')
plt.title('Distribution of Age within Each Country (Percentage)')
plt.xlabel('Country')
plt.ylabel('Age')
plt.legend(title='Percentage')
plt.xticks(rotation=45)
plt.show()
```
